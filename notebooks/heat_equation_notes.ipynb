```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
```

```python
d = 101
x = np.linspace(0, 1, d)
t = np.linspace(0, 1, d)
u = np.zeros((x.shape[0], t.shape[0]))
```

```python
n = 1
L = 1
k = 0.3
xv, tv = np.meshgrid(x, t)

# Initial values
B = (40/(n*np.pi))*(1-(-1)**n)
u = B * np.sin(n*np.pi*xv/L)*np.exp(-k*tv*(n*np.pi/L)**2)

for ni in range(2, 10):
    B = (40/(ni*np.pi))*(1-(-1)**ni)
    u += B * np.sin(ni*np.pi*xv/L)*np.exp(-k*tv*(ni*np.pi/L)**2)
```

```python
fig, ax = plt.subplots(figsize=(7,7))

for i in range(0, d):
    ax.plot(xv[i, :], u[i, :], c='k', alpha=i/d)
ax.set_xlabel('x')
ax.set_ylabel('u')
```

```python
fig, ax = plt.subplots(figsize=(7,7))

for i in range(0,d):
    ax.plot(tv[:, i], u[:, i], c='k', alpha=(1 - (i/d)))
ax.set_xlabel('t')
ax.set_ylabel('u')
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xv, tv, u, cmap=cm.coolwarm)

ax.set_xlabel('x')
ax.set_ylabel('t')
ax.view_init(elev=30, azim=0)
```
